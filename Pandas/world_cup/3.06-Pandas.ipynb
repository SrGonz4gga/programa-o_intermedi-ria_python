#### Q4. 
* Crie um dataframe `gols` com duas colunas `pa√≠s` e `total_gols` com o total de gols marcados por cada pa√≠s durante todas as copas do mundo, juntando gols em casa (`gols_1`) e gols como time visitante (`gols_2`).

#### Q5. 
Qual pa√≠s tomou mais cart√µes amarelos somando todas as copas?

Neste exerc√≠cio voc√™ vai trabalhar com as colunas `cartao_amarelo_1` e `cartao_amarelo_2` onde cada observa√ß√£o √© uma string represetando uma lista dos cart√µes amarelos de um √∫nico jogo na forma `minuto|placar|jogador(a)`. Por exemplo:
```
['16‚Äô|1:0|Rosana G√≥mez', '20‚Äô|2:0|Gabriela Ch√°vez]
```

Recomendo criar colunas `num_cartoes_1` e `num_cartoes_2` a partir de cada coluna `cartao_amarelo_X` usando o m√©todo gen√©rico `apply` para chamar uma fun√ß√£o que voc√™ vai criar para transformar uma observa√ß√£o de cart√£o amarelo em uma contagem de elementos da lista. 

Lembre de levar em considera√ß√£o que muitas observa√ß√µes s√£o `NaN`. 

Em seguida fa√ßa sua m√°gica para agrupar as informa√ß√µes por pa√≠s, acumular os cart√µes de jogos em casa e jogos visitante, e produzir o resultado final como apresentado a seguir (valores inventados):

```
pa√≠s        cart√µes amarelos
Argentina   97
England     93
Australia   93
...
```

````python
def num_cartoes(linha):
    somador = 0
    for c in linha:
        if(c == ','):
            somador +=1
    return somador+1

df_cartoes1 = wc.groupby('time_1')['cartao_amarelo_1'].sum().reset_index()
df_cartoes2 = wc.groupby('time_2')['cartao_amarelo_2'].sum().reset_index()

colunas1_converter = ['time_1', 'cartao_amarelo_1']
colunas2_converter = ['time_2', 'cartao_amarelo_2']

df_cartoes1[colunas1_converter] = df_cartoes1[colunas1_converter].astype("string")
df_cartoes2[colunas2_converter] = df_cartoes2[colunas2_converter].astype("string")

df_cartoes1['cartao_amarelo_1'] = df_cartoes1['cartao_amarelo_1'].apply(num_cartoes)
df_cartoes2['cartao_amarelo_2'] = df_cartoes2['cartao_amarelo_2'].apply(num_cartoes)

df_cartoes1.columns = ['pa√≠s', 'cart√µes amarelos'] 
df_cartoes2.columns = ['pa√≠s', 'cart√µes amarelos'] 


df_cartoes = pd.concat([df_cartoes1, df_cartoes2]).groupby('pa√≠s')['cart√µes amarelos'].sum().reset_index()
df_cartoes = df_cartoes.sort_values(by='cart√µes amarelos', ascending = False).reset_index(drop=True)

display(df_cartoes)



    
````

#### Q6.
Qual o top10 jogadores com mais gols em copa? Considere gols em jogo e gols de p√™nalti.

Para conseguir essa informa√ß√£o, voc√™ precisar√° trabalhar com as colunas:
```
10  gols_1_detalhes         
11  gols_2_detalhes    
14  gols_1_penalti     
15  gols_2_penalti     
```

Essas tamb√©m s√£o colunas textuais, onde cada observa√ß√µes apresenta todos os gols de uma partida separados pelo caracter `|`. Cada gol est√° na forma `'Jogador(a) ¬∑ minuto‚Äô'`. Por exemplo:
```
'Alex Morgan ¬∑ 12‚Äô|Rose Lavelle ¬∑ 20‚Äô'
```

Lembre de levar em considera√ß√£o que muitas observa√ß√µes s√£o `NaN`. 

Recomendo criar um dicion√°rio √† parte, onde cada chave ser√° um jogador encontrado nessas colunas do dataframe `wc` e o valor correspondente ser√° a contagem de ocorr√™ncias desses nomes.

Em seguida basta converter o seu dicion√°rio em um dataframe e imprim√≠-lo na forma (valores n√£o s√£o inventados üòÅ):
```
jogador(a)      num_gols 
Marta           17
Miroslav Klose  16
... 
```

````python
DICIONARIO_JOGADORES = {}

def num_gols_jogador(linha):
    if pd.isna(linha):
        return

    global DICIONARIO_JOGADORES 
    if "|" in linha:
        partes = linha.split("|")
    else:
        partes = [linha]
     
    for p in partes:
        nome = ""
        for i in range(len(p)):
            if (p[i] != "¬∑" and p[i] != "("):
                nome+=p[i]
            else:
                nome = nome.strip()
                if(nome not in DICIONARIO_JOGADORES):
                    DICIONARIO_JOGADORES[nome] = 1
                else:
                    DICIONARIO_JOGADORES[nome] += 1
                break





wc_copia = wc.copy()

wc_copia['gols_1_detalhes'].apply(num_gols_jogador)
wc_copia['gols_2_detalhes'].apply(num_gols_jogador)
wc_copia['gols_1_penalti'].apply(num_gols_jogador)
wc_copia['gols_2_penalti'].apply(num_gols_jogador)


df_goleadores = pd.DataFrame(list(DICIONARIO_JOGADORES.items()),columns=['Jogador(a)', 'num_gols'])

df_goleadores = df_goleadores.sort_values(by='num_gols', ascending=False).reset_index(drop=True)

df_goleadores.index = range(1, len(df_goleadores)+1)
display(df_goleadores)
````
